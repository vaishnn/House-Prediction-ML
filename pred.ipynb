Importing libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

Loading DataSet

```python
data_house = pd.read_csv("dataset/train (1).csv");
data_house.head()
#Checking Data is Loaded perfectly
```

Checking the info of Data Set

```python
data = data_house.copy()
data.describe().T
```

```python
data.info()
```

Removing Column with Missing Value(More than 50%) 

```python
data.isnull().sum().sort_values(ascending=False).iloc[:20,]
```

```python
data.drop(columns=["PoolQC","MiscFeature","Alley","Fence","FireplaceQu","LotFrontage"], inplace=True)
```

```python
data.head()
```

```python
object_col = data.columns
```

Dropping Some More Variable for Reducing OverFitting

```python
data = data.drop(columns=["Street","MSZoning","LotShape","LandContour","RoofMatl","ExterCond","BsmtFinType2","Heating","Functional"])
```

```python
data = data.drop(columns=["Utilities","SaleType","LandSlope"])
```

```python
data.head()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
for col in object_col:
    encoder = LabelEncoder()
    data[col] = encoder.fit_transform(data[col])
```

```python
data.info()
```

Feature Selection

```python
data.head()
```

Treating Missing Values "NA"

```python
plt.figure(figsize=(16,5))
sns.lineplot(x=data["LotArea"],y=data["SalePrice"]);
plt.title("Lot Area VS Sale Price")

plt.figure(figsize=(16,5))
sns.lineplot(x = data["YearBuilt"], y = data["SalePrice"])
plt.title("Year Build VS Sale Price")

plt.figure(figsize=(16,5))
sns.lineplot(x = data["OverallQual"], y = data["SalePrice"])
plt.title("Overall Quality VS Sale Price")

plt.figure(figsize=(16,5))
sns.lineplot(x = data["OverallCond"], y = data["SalePrice"])
plt.title("Overall Condition VS Sale Price")

plt.figure(figsize=(16,5))
sns.lineplot(x = data["MSSubClass"], y = data["SalePrice"])
plt.title("MSSubClass VS Sale Price")

plt.figure(figsize=(16,5))
sns.lineplot(x = data["GrLivArea"], y = data["SalePrice"])
plt.title("GrLivArea VS Sale Price")


```

Choosing the Feature for training the algorithms

So Choosing "OverAllQual" , "GrLivArea" , "OverAllCond" , "YearBuild" for training the Algorithm  

```python
x = data[["OverallQual","GrLivArea","OverallCond","YearBuilt"]]
y = data[["SalePrice"]]
```

Treating Missing DataSet

```python
x.head()
```

Test and Train Split of Data

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.5);
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression

regress_model = LinearRegression()
regress_model.fit(xtrain,ytrain);
```

```python
y_pred = regress_model.predict(xtest)
```

```python
for i in range(y_pred.shape[0]):
    if(y_pred[i] < 0):
        y_pred[i] = 0
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_squared_log_error
mean_squared_log_error(ytest,y_pred)
```

```python
r2_score(ytest,y_pred)
```
